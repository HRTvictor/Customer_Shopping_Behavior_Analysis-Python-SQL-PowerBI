```python
import pandas as pd

df = pd.read_csv("custom.csv")
```

```python
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x : x.fillna(x.median()))
```

```python
#converting all columns string as snack case and also removing special characters.
df.columns = df.columns.str.lower()
```

```python
df.columns = df.columns.str.replace(" " , "_")
```

```python
df = df.rename(columns = {'purchase_amount_(usd)' : "purchase_amount_usd"})
```

```python
df.columns
```

```python
df.head(10)
```

```python
#Here values of 'subscription_status' ,'promo_code_used' and 'discount_applied' are looks same, so let's check 
(df["subscription_status"] == df["discount_applied"]).all()
```

```python
(df["promo_code_used"] == df["discount_applied"]).all()
```

```python
#But only 'promo_code_used' and 'discount_applied' are same ,so I'll remove any one like 'promo_code_used'.
df = df.drop(('promo_code_used'),axis = 1)
```

```python
df.columns
```

```python
#Creating a column "age_group" to group as per age .
labels = ['Young_Adult','Adult','Middle_aged','Senior']
df['age_group'] = pd.qcut(df['age'], q=4 , labels = labels) 
```

```python
df["age_group"].head(10)
```

```python
#Checking Data type of each columns.
df.info()
```

```python
#Checking Duplicate.
df.drop_duplicates(subset = ['customer_id'] ,keep = 'first' ,inplace = True)
df[df.duplicated()].shape
```

```python
#Checking for inconsistent formating.
for col in df.columns:
    if df[col].nunique() <20:
        print(df[col].value_counts())
        print("-"*50)
```

```python
frequency_mapping = {
    'Fortnightly' : 14,
    'Weekly': 7,
    'Annually' : 365,
    'Quarterly' : 90,
    'Every 3 Months' : 90,
    'Bi-Weekly' : 14,
    'Monthly' : 30
}

df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```
